Trying out the codes using fuction parameter

```python
def two_sub(a,b):
    return a-b

a = int(input("Enter your first number : "))
b = int(input("Enter your second number : "))
sub_result = two_sub(a,b)
print(f'Substraction of {a} and {b} is {sub_result}')
```
