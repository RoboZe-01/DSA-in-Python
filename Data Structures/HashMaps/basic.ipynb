## Benefits of Hashmaps

1. Custom keys as earier for engineers to work with 
2. Hashmaps allows for search O(1) , where arrays / linked list O(log(n))

## Let us understand hashmaps with library analogy

Imagine your'e at a massive library and you want to find the book of "Concepts of Physics" by "HC verma"

There is two way you can do this, i.e good way and bad way

- Bad way( array /List): Walking through every asile, checking every single book until you find it -> O(n) time 

- **Good way** ( Hashmap) = You have a magical catalog that tells you : "Books starting with 'T' are on Floor 3, aisle 5, Shelf 2" -> O(1) time

That's exactly what a hashmap does! 




## What is a Hashmap ?

A **hashmap** ( called dict in Python) stores **key-value pairs** . You give it a key, and it instantly gives you the value 

```python
# Creating a hashmap

student_grade = {

    "Prem" : 95,
    "Rahul": 87,
    "Subha" : 90,
    "Om" : 88
}

# Access - O(1) time!
print(student_grade['Om'])  # Output : 88
```
