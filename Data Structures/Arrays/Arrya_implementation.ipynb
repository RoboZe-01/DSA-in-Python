### Adding Elements in the array


- append('value u want to add') : this adds the element at the last of the array

- .insert(index , element ) : insert adds element in specified index

```python
new_arr = arry.array('i',[3,4,5,6,7,8,9,10])
for i in range(0,len(new_arr)):
    print(new_arr[i],end= " ")
print(end="\n")

# Adding elements in the new array using append 

new_arr.append(11)
print(new_arr)

# using insert 

new_arr.insert(0,2)
print(new_arr)

```

## Slicing

Process of getting sequencing from the data which will have starting and ending endex

```python
# Slicing operation in an array

x = list(range(0,100,3))
arr = arry.array('i',x)

# slicing operations
arr0 = arr[10:20]
for i in range(0,len(arr0)):
    print(arr0[i],end=" ")
print("\n")

# Negative indexing

arr1 = arr[10:-10]
for i in range(0,len(arr1)):
    print(arr1[i],end=" ")
print("\n")


# Reversing the order using slicing 
arr2 = arr[::-1]
for i in range(0,len(arr2)):
    print(arr2[i],end=" ")
print("\n")    
```
