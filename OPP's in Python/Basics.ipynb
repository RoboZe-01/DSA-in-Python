## The Blueprint: Classes and Objects

Think of a Class as a blueprint for a house, and an Object as the actual house built from that blueprint.

- Attributes: Data stored inside the object (e.g., color, square footage).

- Methods: Actions the object can perform (e.g., open_door, turn_on_lights).
