## Inherintance In Python

Inheritance is a fundamental concept in Object-Oriented Programming (OOP) that allows a class to inherit attributes and methods from another class .

```python
# Inheritance ( Single inheritance )


## Parent Class
class Car :

    # Parent class
    def __init__(self, windows,doors,enginetype):
        self.windows = windows
        self.doors = doors
        self.enginetype = enginetype

    def drive(self):
        print(f"The person will dive the {self.enginetype}  Car ")

```

```python
car1 = Car(4,5,"Petrol")
print(car1.drive())
```

```python
class Tesla(Car) :
    def __init__(self,windows,doors,enginetype,is_selfdriving):
        super().__init__(windows, doors,enginetype)
        self.is_selfdriving = is_selfdriving
    
    #instance method
    def selfdriving(self):
        print(f"Tesla supports self driving : {self.is_selfdriving} ")

```

 The super method is used to inherit the attributes from the parent class 

```python
tesla1 = Tesla(4,5,"electric",True)
tesla1.selfdriving()
```

```python
### Multiple  Inheritance

## When a class inherits from more than one base class

# Base class 1
class Animal :
    def __init__(self,name):
        self.name = name

    def speak(self):
        print("Subclass must implement this method")    


# Base class 2
class Pet:
    def __init__(self,owner):
        self.owner = owner

# Derived class 
class Dog(Animal,Pet):
    
    def __init__(self,name,owner):
        Animal.__init__(self,name)
        Pet.__init__(self,owner)

    def speak(self):
        return f"{self.name} says woof"   
            
```

```python
# Create an object

dog = Dog("buddy","krish")
print(dog.speak())
print(f"Owner : {dog.owner}")

```
