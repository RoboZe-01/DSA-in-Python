```python
n=5
print("*"*5)
```

```python
# same but using loop 
n = 5
for i in range (0,5):
    print("*",end='')
```

```python
for i in range(n):
    for j in range(n):
        print("*" ,end='')
```

```python
for i in range(n):
    for j in range(n):
        print("*" )
```

```python
for i in range(n):
    for j in range(n):
        print("*" ,end='')
    print()    
```

### Increasing triangle pattern 

```python
for i in range(n):
    for j in range(i+1):
        print('*',end='')
    print()    
```

```python
for i in range(n):
    for j in range(i,n):
        print('*',end='')
    print()    
```
