# **Top 10 Python comprehensions we should be using**

## Basic list comprehension

```python
# doing normally 

values = []

for x in range(10):
    values.append(x)
print(values)    
```

```python
# Doing same using list comprehensions

values = [ x-1 for x in range(10)]
print(values)
```

**Problem** : Get all the even numbers from 0-50

```python
# using loop normally
evens = []
for num in range(0,50):
    if num%2==0:
        evens.append(num)
print(evens)        
            
```

```python
# using list comprehensions 

evens = [number for number in range(0,50) if number%2==0]
print(evens)
```
