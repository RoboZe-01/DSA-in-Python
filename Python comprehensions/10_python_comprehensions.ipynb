# **Top Python comprehensions we should be using**

## Basic list comprehension

```python
# doing normally 

values = []

for x in range(10):
    values.append(x)
print(values)    
```

```python
# Doing same using list comprehensions

values = [ x-1 for x in range(10)]
print(values)
```

**Problem** : Get all the even numbers from 0-50

```python
# using loop normally
evens = []
for num in range(0,50):
    if num%2==0:
        evens.append(num)
print(evens)        
            
```

```python
# using list comprehensions 

evens = [number for number in range(0,50) if number%2==0]
print(evens)
```

```python
# String that start with "a" and end with "y"

options = ['any','albany' , 'apple','world','hello','']
valid_strings = []

for string in options:
    if len(string)<=1:
        continue

    if string[0]!='a':
        continue

    if string[-1]!='y':
        continue

    valid_strings.append(string)

print(valid_strings)    
    
```

```python
valid_str = [

    string
    for string in options
    if len(string)>=2
    if string[0]=='a'
    if string[-1]=='y'
]

print(valid_str)
```

## **Nested List comprehensions**

Flattening two dimensional list ( list of lists)

```python
matrix = [[1,2,3],[4,5,6],[7,8,9]]

flattened = []
for row in matrix:
    for num in row:
        flattened.append(num)

print(flattened)        
```

```python
flat = [num for row in matrix for num in row]
print(flat)
```
