## Binary Search 

Imp tips : array should be **Sorted** to apply binary search

- Binary search is an efficient algorith for finding a target value within a **sorted list** or array by repeatedly dividing tha search space in half

- It reduces the time **complexity to O(log N)** , making it much faster than linear search 


### Working of binary search 

1. Find the middle element of the array
2. Compare the middle ele with targt key , of equal : return index
3. If the key is smaller : search the left half
4. If the key is larger : search the right half
5 . Repeat untill the ele is found or the search is empty
